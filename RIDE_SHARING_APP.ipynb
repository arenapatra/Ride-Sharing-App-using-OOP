#Introducting the Driver class

**Concept**: A class is a blueprint for creating objects. It defines attributes (data) and methods (behaviors).

```python
Class Driver:
    def__ init __(self, name, vehicle, rating = 5.0):
      self.name = name          #public attribute
      self.vehicle = vehicle    #public attribute
      self._rating = rating     #private attribute
      self.is_available = True  #availability status of the driver
```

**Key Points**

#Encapsulation - Getter and Setter

**Concept**: Encapsulation means controlling how attributes are accessed or modified. A "private" attribute (like `__rating`) should ideally have getters and setters to validate changes.

```python
Class Driver:
    def __init__(self, name, vehicle, rating=5.0):
        self.name = name
        self.vehicle = vehicle
        self._rating = rating
        self.is_available = True
#getter for the protected attribute
    def get_rating(self):
            return self._rating
#setter for the protected attribute , with validation
    def set_rating(self, new_rating):
            if 0 <= new_rating <= 5:
                self._rating = new_rating
            else:
                print("Invalid rating. Must be between 0 and 5.")
```

**Key Points**  
- `get_rating()` provides controlled access to `__rating`.  
- `set_rating()` ensures any new rating is within a valid range (0–5).  
- This helps prevent direct manipulation of `__rating` from outside the class.

### **Adding More Behavior - Toggling Availability**  
**Concept**: A class can have any number of methods describing behaviors. Here, we add a method to switch a driver’s availability status.



```python
class Driver:
    def __init__(self, name, vehicle, rating=5.0):
        self.name = name
        self.vehicle = vehicle
        self.__rating = rating
        self.is_available = True

    def get_rating(self):
        return self.__rating

    def set_rating(self, new_rating):
        if 0 <= new_rating <= 5:
            self.__rating = new_rating
        else:
            print("Invalid rating. Must be between 0 and 5.")

    def __str__(self):
        return f"Driver: {self.name}, Vehicle: {self.vehicle}, Rating: {self.__rating}, Available: {self.is_available}"

    # Toggle driver availability on/off
    def toggle_availability(self):
        self.is_available = not self.is_available
```

**Key Points**  
- `toggle_availability()` flips the `is_available` boolean (`True` <-> `False`).  
- This is straightforward, but it shows how methods can modify object attributes in a controlled way.
